Se trabajo sobre la columna fecha:
Los datos estan en formato string, se filtraron los mismos por los que tienen mas de 4 caracteres y se uso  replace para reemplazar por vacios los caracteres que no servian



```python
df[['mes','dia','anio']] = df['fecha'].str.split(' ', expand=True) 
```

```python
df
```

```python
df['dia'] = df.dia.str.replace(',', '')
```

```python
df['mes'] = pd.to_datetime(df['mes'], format= '%B').dt.month
```

```python
df['fecha'] = df['anio'].astype(str)+'-'+df['mes'].astype(str)+'-'+df['dia'].astype(str)
```

```python
df['fecha'].value_counts
```

```python
df.drop(columns=['mes', 'anio', 'dia'], inplace=True)
```

```python
df.head()
```

```python
df['HORA declarada'] = df['HORA declarada'].str.replace(':', '')
```

```python
df['HORA declarada'] = df['HORA declarada'].str.replace('c', '')
```

```python
df['HORA declarada'] = df['HORA declarada'].str.replace('Z', '')
```

```python
df['HORA declarada'] = df['HORA declarada'].str.replace(';', '')
```

```python
df['HORA declarada']= df['HORA declarada'].str.strip()
```

```python
df['hora']=pd.to_datetime(df['HORA declarada'], format='%H%M').dt.time
```

```python
df.head()
```

```python
df.drop(columns=['HORA declarada'], inplace=True)
```

```python
df
```

Se procedio a renombrar las columnas y homogeneizar la letra poniendo todo en minuscula

```python
df.rename(columns={'OperadOR': 'operador', 'PASAJEROS A BORDO': 'pasajeros_a_bordo', 'crew_aboard': 'personal_a_bordo', 'all_aboard': 'total_a_bordo',
           'passenger_fatalities': 'pasajeros_fallecidos', 'crew_fatalities': 'personal_fallecido', 'cantidad de fallecidos': 'total_fallecidos',
           'ac_type': 'tipo_de_avion','Ruta': 'ruta'}, inplace=True)
```

```python
df = df.reindex(columns=['fecha', 'hora', 'operador','tipo_de_avion','ruta','pasajeros_a_bordo','personal_a_bordo','total_a_bordo','pasajeros_fallecidos','personal_fallecido','ground','total_fallecidos']
                 + df.columns.difference(['fecha', 'hora','tipo_de_avion','ruta' ,'operador', 'pasajeros_a_bordo','personal_a_bordo','total_a_bordo','pasajeros_fallecidos','personal_fallecido','ground','total_fallecidos']).tolist())
```

```python
df.head()
```

Finalmente se eliminaron los datos  nulos de cantidad de fallecidos, ya que no sirven para el analisis de la tasa de mortalidad

```python
df.dropna(subset='total_fallecidos',inplace=True)
```
